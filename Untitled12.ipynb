```python
import pandas as pd
```

```python
stores = pd.read_csv("/content/stores.csv")
products = pd.read_csv("/content/products.csv")
customer = pd.read_csv("/content/customers.csv")
promo = pd.read_csv("/content/promotions.csv")
rules = pd.read_csv("/content/loyalty_rules.csv")
header = pd.read_csv("/content/sales_header.csv")
line = pd.read_csv("/content/sales_line_items.csv")
```

```python
import pandas as pd

# Load
merged = line.merge(header, on="transaction_id", how="left")
merged = merged.merge(products, on="product_id", how="left")
merged = merged.merge(stores, on="store_id", how="left")
merged = merged.merge(customer, on="customer_id", how="left")
merged = merged.merge(promo, on="promotion_id", how="left")

rules = pd.read_csv("loyalty_rules.csv")

# CROSS JOIN rules to ALL rows
merged["key"] = 1
rules["key"] = 1
temp = merged.merge(rules, on="key")

# Filter based on threshold
final_merged = temp[temp["total_amount"] >= temp["min_spend_threshold"]]

# Drop key
final_merged.drop("key", axis=1, inplace=True)

final_merged.to_csv("final_with_loyalty_rules.csv", index=False)
```

```python
import pandas as pd
from datetime import datetime

# ==========================================
# 1. PREPARE VALIDATION SETS
# ==========================================
# We load the master files to check if IDs are real ("Referential Integrity")
try:
    # Create sets for fast lookup (O(1) speed)
    valid_customers = set(pd.read_csv("customers.csv")['customer_id'])
    # Handle store_ids as strings to avoid "1.0" vs "1" mismatch
    valid_stores = set(pd.read_csv("stores.csv")['store_id'].astype(str))

    # Load products to check IDs and Prices
    df_prod_master = pd.read_csv("products.csv")
    valid_products = set(df_prod_master['product_id'])
    # Create a price dictionary for Math Checks: {'P001': 10.50, ...}
    product_price_map = dict(zip(df_prod_master['product_id'], df_prod_master['unit_price']))

    print(" Master Data Loaded for Validation.")
except FileNotFoundError:
    print(" WARNING: Master CSVs not found. Skipping Referential Integrity checks.")
    valid_customers = set()
    valid_stores = set()
    product_price_map = {}

# ==========================================
# 2. RUN VALIDATION LOOP
# ==========================================
# final_merged already exists from your code
df = final_merged.copy()

bad_records = []
clean_records = []

today = datetime.today()

print(f" Starting checks on {len(df)} records...")

for idx, row in df.iterrows():
    reason = ""

    # --- LEVEL 1: BASIC CHECKS ---

    # Rule 1: Completeness
    if pd.isna(row.get("transaction_id")) or pd.isna(row.get("product_id")) or pd.isna(row.get("quantity")):
        reason = "Missing important fields"

    # Rule 2: Zero/Negative Quantity
    elif row.get("quantity", 0) <= 0:
        reason = "Invalid quantity (<=0)"

    # Rule 3: Future Date (Time Travel)
    else:
        try:
            txn_date = pd.to_datetime(row["transaction_date"])
            if txn_date > today:
                reason = "Future transaction date"
        except:
            reason = "Invalid transaction date format"

    # Rule 4: Negative Total Amount
    if reason == "" and pd.notna(row.get("total_amount")) and row["total_amount"] < 0:
        reason = "Negative Transaction Amount"

    # --- LEVEL 2: REFERENTIAL INTEGRITY (The "Ghost" Checks) ---

    # Rule 5: Ghost Customer
    if reason == "" and len(valid_customers) > 0:
        if row["customer_id"] not in valid_customers:
            reason = "Ghost Customer (ID not found)"

    # Rule 6: Ghost Store
    if reason == "" and len(valid_stores) > 0:
        # Clean store_id to string (remove decimals if float)
        s_id = str(row["store_id"]).replace('.0', '')
        if s_id not in valid_stores:
            reason = "Invalid Store ID"

    # --- LEVEL 3: BUSINESS LOGIC (The "Math" Check) ---

    # Rule 7: Price Consistency Check
    # Does (Quantity * Unit Price) match the Line Item Amount?
    if reason == "" and row["product_id"] in product_price_map:
        expected_price = product_price_map[row["product_id"]]
        qty = row["quantity"]
        reported_amount = row["line_item_amount"]

        # Calculate expected total
        expected_total = qty * expected_price

        # Allow small difference ($1.00) for floating point rounding errors
        if abs(expected_total - reported_amount) > 1.0:
            reason = f"Math Error: Exp {expected_total:.2f}, Got {reported_amount:.2f}"

    # --- SAVE RESULTS ---
    if reason != "":
        row_dict = row.to_dict()
        row_dict["dq_reason"] = reason
        bad_records.append(row_dict)
    else:
        clean_records.append(row.to_dict())

# ==========================================
# 3. EXPORT FILES
# ==========================================
bad_df = pd.DataFrame(bad_records)
clean_df = pd.DataFrame(clean_records)

bad_df.to_csv("bad_records.csv", index=False)
clean_df.to_csv("clean_final_dataset.csv", index=False)

print("\n Data Quality Pipeline Completed!")
print(f" Bad Records Rejected: {len(bad_df)}")
print(f" Clean Records Saved:  {len(clean_df)}")
```

```python
import pandas as pd
from datetime import datetime, timedelta

# -------------------------------
# Load Good Data (clean dataset)
# -------------------------------
df = pd.read_csv("clean_final_dataset.csv")

# Load customer table to update loyalty points
customers = pd.read_csv("/content/customers.csv")

# -------------------------------
# STEP 1: STANDARD POINTS
# -------------------------------
df["standard_points"] = df["total_amount"] * 1   # 1 point per $1 spent

# -------------------------------
# STEP 2: BONUS POINTS FOR PROMOTIONS
# -------------------------------
df["bonus_points"] = df["promotion_id"].notna().astype(int) * 10  # +10 points if promotion used

# -------------------------------
# STEP 3: TOTAL POINTS EARNED PER TRANSACTION
# -------------------------------
df["points_earned"] = df["standard_points"] + df["bonus_points"]

# -------------------------------
# STEP 4: AGGREGATE POINTS PER CUSTOMER
# -------------------------------
customer_points = df.groupby("customer_id")["points_earned"].sum().reset_index()
customer_points.columns = ["customer_id", "earned_points"]

# -------------------------------
# STEP 5: UPDATE CUSTOMER TABLE TOTAL POINTS
# -------------------------------
customers = customers.merge(customer_points, on="customer_id", how="left")
customers["earned_points"].fillna(0, inplace=True)

customers["updated_total_points"] = (
    customers["total_loyalty_points"].fillna(0) + customers["earned_points"]
)

# -------------------------------
# STEP 6: WRITE UPDATED CUSTOMER FILE
# -------------------------------
customers.to_csv("updated_customers.csv", index=False)

print("Customer Points Updated ✓")

# -------------------------------
# STEP 7: HIGH SPENDERS (Top 10%)
# -------------------------------
threshold = customers["updated_total_points"].quantile(0.90)
high_spenders = customers[customers["updated_total_points"] >= threshold]

high_spenders.to_csv("high_spenders.csv", index=False)

print("High Spenders file created ✓")

# -------------------------------
# STEP 8: AT-RISK CUSTOMERS
# Rule: points > 0 AND no purchase in last 30 days
# -------------------------------
today = datetime.today()
customers["last_purchase_date"] = pd.to_datetime(customers["last_purchase_date"], errors="coerce")

customers["days_since_last_purchase"] = (today - customers["last_purchase_date"]).dt.days

at_risk = customers[
    (customers["updated_total_points"] > 0) &
    (customers["days_since_last_purchase"] >= 30)
]

at_risk.to_csv("at_risk_customers.csv", index=False)

print("At-Risk Customers file created ✓")

print("\nAll Outputs Generated:")
print("✓ updated_customers.csv")
print("✓ high_spenders.csv")
print("✓ at_risk_customers.csv")
```

```python
import pandas as pd
from datetime import datetime

# ---------------------------------------------------
# STEP 1: Load the cleaned, good dataset
# ---------------------------------------------------
df = pd.read_csv("clean_final_dataset.csv")

# Ensure date column is datetime
df["transaction_date"] = pd.to_datetime(df["transaction_date"], errors='coerce')

# Drop rows with invalid dates
df = df.dropna(subset=["transaction_date"])

# ---------------------------------------------------
# STEP 2: Prepare reference date for Recency
# ---------------------------------------------------
today = datetime.today()

# ---------------------------------------------------
# STEP 3: Calculate RFM components
# ---------------------------------------------------

# Recency = days since last purchase
recency_df = df.groupby("customer_id")["transaction_date"].max().reset_index()
recency_df["Recency"] = (today - recency_df["transaction_date"]).dt.days

# Frequency = total number of transactions
frequency_df = df.groupby("customer_id")["transaction_id"].nunique().reset_index()
frequency_df.rename(columns={"transaction_id": "Frequency"}, inplace=True)

# Monetary = total spending
monetary_df = df.groupby("customer_id")["total_amount"].sum().reset_index()
monetary_df.rename(columns={"total_amount": "Monetary"}, inplace=True)

# ---------------------------------------------------
# STEP 4: Merge all RFM metrics
# ---------------------------------------------------
rfm = recency_df.merge(frequency_df, on="customer_id")
rfm = rfm.merge(monetary_df, on="customer_id")

# ---------------------------------------------------
# STEP 5: Save RFM table
# ---------------------------------------------------
rfm.to_csv("customer_rfm_metrics.csv", index=False)

print("RFM Calculation Completed Successfully!")
print("Output File: customer_rfm_metrics.csv")

# Display sample
rfm.head()
```
